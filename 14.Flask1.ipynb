```python
# Q3. What is App routing in Flask? Why do we use app routes?
```

```python
# Solution 3>

# In Flask web applications, app routing is the mechanism that establishes a connection between specific URLs (web addresses) and the corresponding
# Python functions that handle those requests. It essentially maps URLs to actions within application.

# We use app routes to:

# Organize our code for better maintainability.
# Create meaningful URLs that users can understand.
# Handle different actions (like viewing vs. submitting data) using HTTP methods.
# Build flexible URLs with placeholders for dynamic content.
# App routes are essential for building well-structured and user-friendly Flask web applications.
```
